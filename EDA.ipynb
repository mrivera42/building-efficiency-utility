```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset 
df = pd.read_csv('energy_efficiency_data.csv')
df.head()
```

```python
df.shape
```

```python
# get list of features 
headers = list(df.columns)
headers
```

```python
# look at distribution of each feature 
fig = plt.figure(figsize=(10,10))

for i in range(len(headers)):
    
    feature = headers[i]
    x = df[feature]
    plt.subplot(5,2,i+1)
    plt.hist(x)
    plt.title(feature)
plt.tight_layout()
```

```python
# look at correlation of each feature with heating_load 
df.corr()['Heating_Load'].sort_values().plot(kind='bar')
```

```python
# looks at correlation of each feature with cooling_load
df.corr()['Cooling_Load'].sort_values().plot(kind='bar')
```

```python
# look at how each feature is correlated with eachother 
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(),annot=True, cbar=False, cmap='Blues', fmt='.1f')
```
